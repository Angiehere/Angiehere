Captura de imagenes para formar data set

```python
#librerias
import cv2
import mediapipe as mp
import csv
import os
```

```python
#llamar a mp hands,se delcara que el numero maximo de manos detectadas por imagen es 1
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(max_num_hands=1, min_detection_confidence=0.5)

#ubicacion de las muestras
C0 = r"C:\Users\avale\OneDrive\Escritorio\piaia\Manos pose para dibujar\C0" #No dibujar
C1 = r"C:\Users\avale\OneDrive\Escritorio\piaia\Manos pose para dibujar\C1" #Dibujar

#se crea un data set para el procesamiento de los datos
ARCHIVO = "dataset.csv"
```

```python
#configuracion para el procesamiento de las imagenes
def procesar_carpeta(carpeta, clase):
    muestras = 0
    for archivo in os.listdir(carpeta):
        if not archivo.lower().endswith(('.png', '.jpg', '.jpeg', '.jfif', '.webp')):
            continue
        ruta = os.path.join(carpeta, archivo)
        frame = cv2.imread(ruta)
        if frame is None:
            print(f"No se pudo leer: {archivo}")
            continue
        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        result = hands.process(rgb)
        if result.multi_hand_landmarks:
            lm = result.multi_hand_landmarks[0].landmark
            row = [coord for point in lm for coord in (point.x, point.y, point.z)]
            row.append(clase)
            with open(ARCHIVO, 'a', newline='') as f:
                csv.writer(f).writerow(row)
            muestras += 1
            print(f"Clase {clase} - {archivo}: OK")
        else:
            print(f"Clase {clase} - {archivo}: no se detectó mano, ignorando")
    return muestras

if os.path.exists(ARCHIVO):
    os.remove(ARCHIVO)

print("Procesando clase 0...")
n0 = procesar_carpeta(C0, 0)

print("\nProcesando clase 1...")
n1 = procesar_carpeta(C1, 1)

print(f"\nDataset generado:")
print(f"  Clase 0: {n0} muestras")
print(f"  Clase 1: {n1} muestras")
print(f"  Total:   {n0 + n1} muestras en {ARCHIVO}")

hands.close()
```
